```python
def solution(numbers):
    answer = ''
    max_element = -1
    biggest = []
    
    while(len(numbers) > 0):
        for element in numbers:
            if str(element)[0] > str(max_element)[0]:
                max_element = element
            elif str(element)[0] == str(max_element)[0]:
                if len(str(element)) >= 2:
                    if str(element)[1] > str(max_element)[1]:
                        max_element = element
                    elif str(element)[1] == str(max_element)[1]:
                        if len(str(element)) >= 3:
                            if str(element)[2] > str(max_element)[2]:
                                max_element = element              
        
        biggest.append(max_element)
        numbers.remove(max_element)
        
    for num in biggest:
        answer += str(num)
    return answer

print(solution([6, 10, 2]))
```
