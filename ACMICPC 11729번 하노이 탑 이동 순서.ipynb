```python
def move(num, startPoint, temp, destination):
    if num == 1 :
        print(startPoint, destination)
    else :
        move(num - 1,startPoint,destination, temp)
        move(1, startPoint,temp,destination)
        move(num - 1, temp,startPoint, destination)
```

```python
K = int(input())
print(2**K - 1)
move(K, 1, 2, 3)
```
