```python
def isPrime(number) :
    if number == 1 :
        return False
    elif number == 2 :
        return True
    for i in range(2, number):
        if number % i == 0 :
            return False
    return True
```

```python
n = int(input())
List = list(map(int, input().split()))
count = 0
for number in List :
    if isPrime(number):
        count += 1
print(count)
```
