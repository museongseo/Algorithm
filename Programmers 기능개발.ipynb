input 값은 초기 작업 진도 progress 와 각 기능별 작업 속도 speeds 이다. progress 와 speeds 를 이용하여 모든 기능이 개발되는데 각각 며칠이 걸리는지 계산하는걸 우선 순위로 한다. <br>

이제 times 라는 리스트가 생성이 되었는데 index 가 낮은 위치에 존재하는 기능이 먼저 개발되어야 후속되는 기능이 개발될 수 있기 때문에 첫 원소를 피벗으로 두고 그 원소보다 큰 값이 나오기 전까지 원소의 개수를 세리고 answer 배열에 append 를 할 것이다.

```python
import math

def solution(progresses, speeds):
    times = []
    answer = []
    for i in range(len(progresses)):
        times.append(math.ceil((100 - progresses[i]) / speeds[i]))
    max = times[0]
    count = 1
    for j in range(1, len(times)):
        if (max < times[j]):
            answer.append(count)
            max = times[j]
            count = 1
        else:
            count += 1
    answer.append(count)
    return answer
```
