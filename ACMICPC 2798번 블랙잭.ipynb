```python
N, M = map(int, input().split())
sum = 0
max = 0
num = list(map(int, input().split()))
for i in range(0, N) :
    for j in range(i+1, N):
        for k in range(j+1, N):
            sum = num[i] + num[j] + num[k]
            if ((sum > max) and (sum <= M)) :
                max = sum
print(max)
```
