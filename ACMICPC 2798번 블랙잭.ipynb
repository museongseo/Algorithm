```python
N, M = map(int, input().split())
sum = 0
max = 0
num = []
for i in range(0, N) :
    num[i] = map(int, input().split)
for j in range(0, N) :
    for k in range(j+1, N):
        for l in range(k+1, N):
            sum = num[j] + num[k] + num[l]
            if ((sum > max) and (sum < M)) :
                max = sum
print(max)
```
