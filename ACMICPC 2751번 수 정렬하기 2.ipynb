시간 복잡도가 O(nlogn) 인 알고리즘은 insert sort 를 제외한 comparison sort 이다.

```python
def Partition(A, p, r):
    x = A[r]
    i = p-1
    for j in range(p, r):
        if A[j] <= x:
            i = i + 1
            temp = A[i]
            A[i] = A[j]
            A[j] = temp
    temp = A[i+1]
    A[i+1] = A[r]
    A[r] = temp
    return i+1

def QuickSort(A, p, r):
    if (p < r) :
        q = Partition(A, p, r)
        QuickSort(A, p, q-1)
        QuickSort(A, q+1, r)
        
N = int(input())
A = []
for i in range(N):
    A.append(int(input()))
QuickSort(A, 1, N)
```
