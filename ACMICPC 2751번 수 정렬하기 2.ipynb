시간 복잡도가 O(nlogn) 인 알고리즘은 insert sort 를 제외한 comparison sort 이다.

```python
import sys

def Partition(A, p, r):
    x = A[r]
    i = p-1
    for j in range(p, r):
        if A[j] <= x:
            i = i + 1
            A[i], A[j] = A[j], A[i]
    A[i+1], A[r] = A[r], A[i+1]
    return i+1

def QuickSort(A, p, r):
    if (p < r) :
        q = Partition(A, p, r)
        QuickSort(A, p, q-1)
        QuickSort(A, q+1, r)
        
N = int(input())
A = []
for i in range(N):
    A.append(int(sys.stdin.readline().rstrip()))
QuickSort(A, 0, N-1)
for elements in A:
    print(elements)
```
