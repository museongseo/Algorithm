시간 복잡도가 O(nlogn) 인 알고리즘은 insert sort 를 제외한 comparison sort 이다.

QuickSort 를 구현해봤는데, 평균 시간복잡도는 O(nlogn) 이지만 최악의 경우 O(n^2) 가 나오게 된다. <br>
이 점이 문제에서 계속 실패하게 만드는건가 싶기도 하다..

```python
def QuickSort(A):
    if len(A) <= 1:
        return A
    
    left = []
    right = []
    pivot = A[0]
    
    for i in range(1, len(A)):
        if pivot > A[i]:
            left.append(A[i])
        else:
            right.append(A[i])
    
    return QuickSort(left) + [pivot] + QuickSort(right)

N = int(input())
A = []
for i in range(N):
    num = int(input())
    A.append(num)

sortedA = QuickSort(A)

for element in sortedA:
    print(element)
```
