```python
def MergeSort(arr):
    if len(arr) < 2:
        return arr

    mid = len(arr) // 2
    low_arr = MergeSort(arr[:mid])
    high_arr = MergeSort(arr[mid:])

    sortedA = []
    l = h = 0
    while l < len(low_arr) and h < len(high_arr):
        if low_arr[l] < high_arr[h]:
            sortedA.append(low_arr[l])
            l += 1
        else:
            sortedA.append(high_arr[h])
            h += 1
    sortedA += low_arr[l:]
    sortedA += high_arr[h:]
    return sortedA

N = int(input())
A = []
for i in range(N):
    num = int(input())
    A.append(num)

List = MergeSort(A)

for element in List:
    print(element)
```
