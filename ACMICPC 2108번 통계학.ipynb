```python
def ArithmeticMean(List):
    sum = 0
    for number in List:
        sum += number
    return round(sum/len(List))
```

```python
def Median(List):
    List = sorted(List)
    return List[len(List)//2]
```

```python
def Mode(List):
    List = sorted(List)
    modes = []
    modeDict = dict(zip(List, [0]*len(List)))
    for number in List:
        modeDict[number] += 1
    for key, value in modeDict.items():
        if value == max(modeDict.values()):
            modes.append(key)
    if (len(modes) >= 2) : return modes[1]
    else : return modes[0]
```

```python
def Range(List):
    List = sorted(List)
    max = List[len(List)-1]
    min = List[0]
    return max - min
```

```python
N = int(input())
num = []
for _ in range(0, N):
    num.append(int(input()))
print(ArithmeticMean(num))
print(Median(num))
print(Mode(num))
print(Range(num))
```
