```python
N, M = map(int, input().split())
List = []
for _ in range(N):
    List.append(int(input()))
DPTable = [10001] * (M + 1)
DPTable[0] = 0

for i in range(N):
    for j in range(List[i], M + 1):
        if DPTable[j - List[i]] != 10001:
            DPTable[j] = min(DPTable[j], DPTable[j - List[i]] + 1)
        
if DPTable[M] == 10001:
    print(-1)
else:
    print(DPTable[M])
```
