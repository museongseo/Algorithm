```python
def numberOfResident(k,n):
    count = 0
    if k == 0 :
        return n
    else :
        for i in range(1, n+1):
            count += numberOfResident(k-1,i)
    return count
```

```python
T = int(input())
for j in range(T):
    k = int(input())
    n = int(input())
    print(numberOfResident(k,n))
```
