```python
# career fair 동안 주최할 수 있는 최대 presentation 의 수를 결정하는 것이 문제이다.

def maxEvents(arrival, duration):
    #dictionary 만들기
    #arrival 을 key 값으로, duration 을 item 값으로 하는 dictionary 를 만들 것이다. 중복된 key 값이 존재하면 안 되기 때문에 같은 arrival 값을 가진 원소들에 대해 가장 효율적인 duration 값을 가진 원소들만 남겨두어야하는데, duration 이 가장 짧은 event 를 선택하는 것이 최대 presentation 을 주최하기에 유리하다.

    for i in range(len(arrival)):
        for j in range(i+1, len(arrival)):
            if arrival[i] == arrival[j]:
                if duration[i] >= duration[j]:
                    arrival[i] = -1
                    duration[i] = -1
                else:
                    arrival[j] = -1
                    duration[j] = -1
    
    dict_events = dict(zip(arrival, duration))
    dict_events = sorted(dict_events.items()) #arrival time 에 대해서 dictionary 정렬
    print(dict_events)
    if (-1, -1) in dict_events:
        dict_events.remove((-1, -1))
    arrival = []
    duration = []
    for i in range(0, len(dict_events)):
        arrival.append(dict_events[i][0])
        duration.append(dict_events[i][1])
    print(arrival, duration)
    
    count = 1 #가장 마지막에 수행되는 presentation 의 경우는 비교할 대상이 없이 항상 수행이 가능하므로 1개의 event 를 더 세려준다.
    isPivot = 0
    for i in range(len(arrival)):
        if isPivot == 0:
            pivot_arr = arrival[i]
            pivot_dur = duration[i]
            isPivot = 1
        if pivot_arr != arrival[i]:
            if pivot_arr + pivot_dur <= arrival[i] + duration[i]: #도착 시간이 더 빠르더라도, 수행시간이 너무 길어 도착시간이 더 느린 event 가 빠른 시간에 끝날 수 있는
경우도 존재한다. (ex. (arr, dur) 가 (1,5) 와 (2,1) 와 같은 경우)
                if pivot_arr + pivot_dur <= arrival[i]:
                    count += 1
                    pivot_arr = arrival[i]
                    pivot_dur = duration[i]
            else:
                pivot_arr = arrival[i]
                pivot_dur = duration[i]
        return count
```
