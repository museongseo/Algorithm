```python
def solution(phone_book):
    phone_book.sort(reverse = True) #길이가 가장 긴 번호가 앞으로 오도록 정렬
    for i in range(len(phone_book) - 1):
        for j in range(len(phone_book[i]) - 1):
            if (phone_book[i])[0:(len(phone_book[i])-j-1)] in phone_book:
                return False
    answer = True
    return answer
```

210809) 길이가 가장 긴 번호를 앞으로 오도록 정렬해서, 가장 긴 번호부터 뒤에서 몇 개의 숫자를 잘라내어 동일한 값을 가진 전화번호가 전화번호 목록에 있는지 확인하는 과정을 거친다. 목록에 있다면 그 전화번호가 접두사가 되는 것이므로 False 를 반환한다. 이렇게 코드를 작성하니 정확성 테스트는 모두 통과했는데 효율성 테스트의 3, 4번 문제에 대하여 통화하지 못했다. 그래고 Key 와 Value 를 사용하는 해시 형태로 코드를 구현하지 못했다.
