```python
def solution(priorities, location):
    sequences = [i for i in range(len(priorities))]
    prints = []
    while priorities:
        p_top = priorities.pop(0)
        s_top = sequences.pop(0)
        if not priorities:
            prints.append(s_top)
            break
        if max(priorities) > p_top:
            priorities.append(p_top)
            sequences.append(s_top)
        else:
            prints.append(s_top)
    answer = prints.index(location) + 1
    return answer
```
