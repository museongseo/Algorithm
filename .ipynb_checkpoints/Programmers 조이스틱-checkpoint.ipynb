```python
def solution(name):
    answer = 0
    name = list(name)
    
    #상하 컨트롤의 수
    for i in range(len(name)):
        print(min(ord(name[i]) - 65, 91 - ord(name[i])))
        print(ord('N'))
        answer += min(ord(name[i]) - 65, 91 - ord(name[i]))
```

상하 컨트롤의 수를 구하는 것은 위와 같이 간단하게 구할 수 있다.

좌우 컨트롤의 수를 구하는 것이 이 문제의 핵심인데, 문자가 a개의 알파벳 + b개의 연속된 A + c개의 알파벳으로 구성되어있을 때, a + b + c - 1 보다 (a-1) * 2 + c 가 더 작은 경우에는 index 가 0일때부터 쭉 가다가 'A'를 만나면 다시 index 가 0 인 지점까지 돌아와서 -1 부터 연속된 A가 끝날 때까지 가야한다.
