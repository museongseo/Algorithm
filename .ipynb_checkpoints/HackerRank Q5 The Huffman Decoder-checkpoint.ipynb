```python
# Huffman codes 를 decoding 하는 것이 문제이다.

def decode(codes, encoded):
    answer = []
    
    #codes dictionary 생성
    dict_codes = {} #codes 에 있는 값을 각각 dictionary 의 key 와 value 로 변환한다.
    
    for i in range(len(codes)):
        if codes[i].split('\t')[0] == None:
            continue
        dict_codes[codes[i].split('\t')[0]] = codes[i].split('\t')[1]

    #encoded 에서 Huffman codes 에 해당하는 값 추출
    i = 0
    for j in range(len(encoded)+1):
        if encoded[0:i] in dict_codes.values():
            for key, value in dict_codes.items():
                if value == encoded[0:i]:
                    if key == '[newline]':
                        answer.append("\n")
                    elif key == ' ':
                        answer.append("\t")
                    else:
                        answer.append(key)
            encoded = encoded[i:]
            i = 0
        i += 1
    answer = ''.join(answer)
    return answer
```
