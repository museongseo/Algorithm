```python
# 주어진 string 에서 모든 하위 substring 을 제거하는 것이 문제이다.
# 간단하게 생각해서 string 안에 있는 모든 중복된 알파벳을 제거한다고 생각했다.
def getMinDeletions(s):
    list_s = list(s) # string 을 각 알파벳을 원소로 가진 list 로 바꾸고, list 안에 있는 중복된 원소를 제거하기로 했다.
    new_list_s = []
    for x in list_s: # string 의 순서를 지킨 채로 중복된 알파벳을 제거하려고 했는데, 문제에서 물어보는게 단순히 제거를 수행해야할 횟수만을 물어보는 것이기 때문에 사실 순서를 지키지 않아도 상관없었을 것 같다.
if x not in new_list_s:
    new_list_s.append(x)
    new_s = "".join(new_list_s)
    return len(s) - len(new_s)
```
