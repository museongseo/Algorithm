```python
def solution(numbers):
    string = []
    for num in numbers:
        string.append(str(num))
    string.sort(key=lambda x: x*3, reverse=True)
    answer = ''
    for num in string:
        answer += num
    return answer
```

```python
def solution(numbers):
    string = []
    for num in numbers:
        string.append(str(num))
    string.sort(key=lambda x: x*3, reverse=True)
    answer = str(int(''.join(string)))
    return answer
```
