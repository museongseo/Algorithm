```python
def d(num):
    sum = num + num//1000 + ((num//100)-(num//1000 * 10)) + (num//10)-(num//100 * 10) + (num%10)
    if (sum <= 10000):
        return sum
    else: return 0

isself = [0 for i in range(20000)]

for j in range(10001):
    isself[d(j)] += 1
    
for k in range(10001):
    if (isself[k] == 0):
        print(k)
```
