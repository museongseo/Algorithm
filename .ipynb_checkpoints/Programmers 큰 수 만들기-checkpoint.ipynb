```python
def solution(number, k):
    answer = ''
    max = 0
    while k > 0 :
        for num in number[0:-k] :
            if int(num) > max :
                max = int(num)
        answer += num
        k -= 1
    return answer
```

210727) number 의 길이가 n 이고 제거할 수의 개수가 k 일 때, 첫 번째 수는 제일 뒤에서 k-1 개를 제외한 나머지 수들 중 가장 큰 수를 선택한다. 가장 큰 수가 여러 개라면 그 중 제일 앞에 있는 수를 선택한다.

```python
def solution(number, k):
    answer = ''
    max = 0
    index = 0
    while k > 0 :
        for num in number[index:-k] :
            if int(num) > max :
                max = int(num)
        answer += str(max)
        k -= 1
        index = number.index(str(max))
    return answer
```
