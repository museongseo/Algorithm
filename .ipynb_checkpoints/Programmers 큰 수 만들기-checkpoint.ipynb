```python
def solution(number, k):
    answer = ''
    max = 0
    while k > 0 :
        for num in number[0:-k] :
            if int(num) > max :
                max = int(num)
        answer += num
        k -= 1
    return answer
```

210727) number 의 길이가 n 이고 제거할 수의 개수가 k 일 때, 첫 번째 수는 제일 뒤에서 k-1 개를 제외한 나머지 수들 중 가장 큰 수를 선택한다. 가장 큰 수가 여러 개라면 그 중 제일 앞에 있는 수를 선택한다.

210729) 스택을 이용해보자. 제일 앞자리의 수가 가장 큰 수가 되도록 찾는 부분해가 결국 전체의 해가 된다. stack 에 number 의 제일 앞에 있는 값을 넣는다. k가 0보다 크다면, number 의 다음 순서에 위치하는 숫자와 비교를 한 후, 더 큰 값만 남겨둔다. 반복이 한 번 될 때마다 k를 하나씩 감소시켜준다. k가 0이 되었을 때, stack 안에 존재하는 값이 가장 큰 수가 된다.
