```python
def Sort(list):
    for i in range(len(list)-1, 0, -1):
        for j in range(0, i):
            if list[j] > list[j+1]:
                temp = list[j+1]
                list[j+1] = list[j]
                list[j] = temp
```

```python
N = int(input())
num = []
for _ in range(0, N):
    num.append(int(input()))
Sort(num)
for number in num :
    print(number)
```
