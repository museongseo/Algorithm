```python
# 주어진 tickets array 에서 연속되거나 같은 정수의 값을 가진 subsequence 의 길이의 최대값을 찾는 것이 문제이다.
# Complete the maxTickets function below.
def maxTickets(tickets):
    tickets = sorted(tickets) #내림차순으로 tickets array를 정렬한다.
    count = 1
    max_count = 0
    pivot = -1

for i in range(len(tickets)):
    if (pivot == -1):
        pivot = tickets[i]

    elif (abs(pivot - tickets[i]) <= 1): # pivot - tickets[i] 의 값에 절댓값을 취한 값이 1 이하이면 subsequence 에 tickets[i] 를 append 한다.
        pivot = tickets[i]
        count += 1

    else:
        if (max_count <= count):
            max_count = count
            pivot = tickets[i]
            count = 1
return (max_count)
```
