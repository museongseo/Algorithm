```python
def solution(scoville, K):
    count = 0
    while min(scoville) < K :
        scoville.sort()
        scoville.append(scoville.pop(0) + 2 * scoville.pop(0))
        count += 1
    return count
```

```python
import heapq

def solution(scoville, K):
    count = 0
    heap = []
    mix = 0
    for num in scoville:
        heapq.heappush(heap, num)
    while heap[0] < K :
        heapq.heappush(heap, heapq.heappop(heap) + 2 * heapq.heappop(heap))
        count += 1
    return count
```
