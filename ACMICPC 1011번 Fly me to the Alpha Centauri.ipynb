```python
def Teleport(x, y) :
    distance = y - x
    count = 0
    while (1) :
        if distance <= (count**2) :
            if distance > (count**2 - count) : 
                Teleport_Count = 2 * count - 1
                return Teleport_Count
            else :
                Teleport_Count = 2 * (count - 1)
                return Teleport_Count
        else :
            count += 1
```

```python
T = int(input())
for j in range(T):
    x, y = map(int, input().split())
    print(Teleport(x, y))
```
